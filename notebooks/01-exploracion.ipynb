#### 3.2 Nulos

```python
nulos = df.isnull().sum()
porcentaje = (nulos / len(df) * 100).round(2)

pd.DataFrame({
    "nulos": nulos,
    "porcentaje %": porcentaje
})
```

#### 3.3 Ejemplos de cada columna

```python
for col in df.columns:
    print(f"\n{'='*40}")
    print(f"  {col.upper()}")
    print(df[col].dropna().head(5).tolist())
```

## Conclusiones - Qué hay que limpiar

### precio
- Formato: "USD 530.000"
- Limpiar: sacar "USD " y los puntos → int

### expensas
- Formato: "$ 770.000 Expensas"
- Limpiar: sacar "$ " y " Expensas" y los puntos → int
- 46% nulos → rellenar con 0

### metros
- Formato: "172 m² tot."
- Limpiar: sacar " m² tot." → int
- 18 nulos → eliminar filas

### ambientes
- Formato: "4 amb."
- Limpiar: sacar " amb." → int
- 9% nulos → analizar si imputar o eliminar

### banos
- Formato: "3 baños" o "1 baño"
- Limpiar: sacar " baños" y " baño" → int
- 7.5% nulos → analizar si imputar o eliminar

### direccion
- Formato inconsistente, no la usaremos en el modelo
- La dejamos como referencia

### barrio
- Formato: "Recoleta, Capital Federal" o "Las Cañitas, Palermo"
- Limpiar: quedarnos solo con el barrio principal
- Dos formatos distintos → hay que manejar ambos
