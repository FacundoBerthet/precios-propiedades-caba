### Nulos

```python
# Nulos por columna
df.isnull().sum()
```

```python
# Porcentaje de nulos por columna
nulos = df.isnull().sum()
porcentajes = (nulos / len(df) * 100).round(2)

print("Porcentajes de nulos por columna:")
print(porcentajes)
```
